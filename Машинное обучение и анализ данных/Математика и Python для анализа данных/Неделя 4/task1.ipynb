## Импорт библиотек

```python
import pandas as pd
import numpy as np
import scipy.stats as sts
import matplotlib.pyplot as plt
```

## Генерация выборки и рассчет теоритеческой плотности вероятности

```python
ex = sts.expon() 
sample = ex.rvs(1000)

# теоритеческая плотность вероятности
x = np.linspace(0, 15, 1000)
pdf = ex.pdf(x)
```

## Отрисовка графика

```python
plt.plot(x, pdf)
plt.hist(sample, density=True, edgecolor='black')
plt.xlabel('X')
plt.ylabel('Func')
```

## Создание функции для отрисовки графика выборочных средних случайной величины и теоретической оценки плотности вероятности для распределения
### Входной параметр n - объем выборки

```python
def create_pdf(n): 
    # Выборочные средние по выборкам
    means = [np.mean(ex.rvs(n)) for _ in range(1000)]
    
    # теоритеческая плотность вероятности для нормального распределения
    mu, sigma = 1, (1./n) ** 0.5
    norm_rv = sts.norm(loc=mu, scale=sigma)
    x = np.linspace(0,4,100)
    pdf = norm_rv.pdf(x)
    
    # построение графика
    plt.plot(x, pdf, linewidth = 2.5)
    plt.hist(means, density=True, edgecolor='black')
    plt.title('n = %d' % n)
    plt.xlabel('X')
    plt.ylabel('Func')
```

```python
create_pdf(3)
```

```python
create_pdf(5)
```

```python
create_pdf(10)
```

```python
create_pdf(50)
```

## Вывод: чем больше объем выборки, тем ближе распледеление приближается к нормальному.
